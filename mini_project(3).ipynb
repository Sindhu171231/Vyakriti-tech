```python
# Load dataset
data = pd.read_csv('customer_churn.csv')

# Check the first few rows to ensure it loaded correctly
print(data.head())
```

```python
# Independent variable (X) and dependent variable (y)
X = data[['tenure']]  # tenure as independent variable
y = data['MonthlyCharges']  # MonthlyCharges as dependent variable
print(X)
print(y)
```

```python
# Split the dataset into train and test sets (70:30 ratio): Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(X_train, X_test, y_train, y_test )
```

```python
#Build the model on train set and predict the values on test set:
# Initialize the linear regression model
model = LinearRegression()

# Train the model on the training set
model.fit(X_train, y_train)
```

```python
# Predict the MonthlyCharges on the test set
y_pred = model.predict(X_test)
print(y_pred)

```

```python
# Calculate the RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"Root Mean Square Error: {rmse}")
```

```python
# Find the prediction errors and store them in 'error'
error = y_test - y_pred
print(error.head())
```
